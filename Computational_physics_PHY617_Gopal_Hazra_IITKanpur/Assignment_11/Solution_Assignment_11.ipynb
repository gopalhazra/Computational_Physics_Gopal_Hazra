```python
# Question-1

def harmonic_oscillator(x, V0, a):
    return V0*x**2/a**2
def anharmonic_oscillator(x, V0, a):
    return V0*x**4/a**4
def schrodinger_equation(x, psi, psi_prime, V, E, m, h_bar):
    psi_double_prime=2*m*(V-E)*psi/h_bar**2
    return psi_prime, psi_double_prime
def solve(x, V, m, h_bar, E_guess):
    psi=np.zeros_like(x)
    psi_prime=np.zeros_like(x)
    psi[0]=0
    psi_prime[0]=1
    for i in range(1, len(x)):
        dx=x[i]-x[i-1]
        psi_prime[i], psi[i]=schrodinger_equation(x[i-1], psi[i-1], psi_prime[i-1], V[i-1], E_guess, m, h_bar)
        psi_prime[i]*=dx
        psi[i]*=dx
    return psi
m=9.109e-31
h_bar=1.054e-34
V0=50*1.6e-19
a=1e-11

# Part-1
x=np.linspace(-10*a, 10*a, 1000)
energies_harmonic=[]
for n in range(3):
    E_guess=(n+0.5)*h_bar**2/(2*m*a*a)
    psi=solve(x, harmonic_oscillator(x, V0, a), m, h_bar, E_guess)
    energies_harmonic.append(E_guess)
    print(f"Energies of state {n+1} for harmonic oscillator comes out to be {E_guess/1.6e-19}")
print(f"The energy spacing = {(energies_harmonic[-1]-energies_harmonic[-2])/1.6e-19} eV")

# Part-2
energies_anharmonic=[]
for n in range(3):
    E_guess=((n+0.5)-0.375*(n+0.5)**2)*h_bar**2/(2*m*a*a)
    psi=solve(x, harmonic_oscillator(x, V0, a), m, h_bar, E_guess)
    energies_anharmonic.append(E_guess)
    print(f"Energies of state {n+1} for anharmonic oscillator comes out to be {E_guess/1.6e-19}")
```

```python
# Question-2

A=np.array([[0, 0, 2, 1, 2], 
            [0, 1, 0, 2, -1], 
            [1, 2, 0, -2, 0], 
            [0, 0, 0, -1, 1], 
            [0, 1, -1, 1, -1]], float)
v=np.array([1,1,-4,-2,-1], float)
n=len(v)
for i in range(n):
    x=A[i][i]
    for j in range(i+1, n):
        y=A[i][j]
        if x==0 and y!=0:
            A[[i,j]]=A[[j,i]]
            v[[i,j]]=v[[j,i]]
            break
for i in range(n):
    ele=A[i][i]
    A[i,:]=A[i]/ele
    v[i]=v[i]/ele
    for j in range(i+1, n):
        mul=A[j][i]
        A[j,:]=A[j,:]-mul*A[i,:]
        v[j]=v[j]-mul*v[i]
X=np.zeros(n)
for index in range(n-1, -1, -1):
    X[index]=v[index]
    for i in range(index+1, n):
        X[index]=X[index]-A[index][i]*X[i]
ans=X.reshape(-1, 1)
print(ans)
```

```python
# Question-3

A=np.array([[3, -1, 4], [-2, 0, 5], [7, 2, -2]], float)
v=np.array([[6, -4], [3, 2], [7, -5]], float)

# LU decomposition function
def Dec(a):
    n=len(a)
    l=np.eye(n)
    u=a.copy()
    for i in range(n):
        for j in range(i+1, n):
            fac=u[j,i]/u[i,i]
            l[j,i]=fac
            u[j,:]-=fac*u[i,:]
    return l, u
    
# Forward substitution
def fsub(l, v):
    n=len(l)
    y=np.zeros((n, v.shape[1]))
    for i in range(n):
        y[i,:]=v[i,:]
        for j in range(i):
            y[i,:]-=l[i,j]*y[j,:]
    return y
    
# Backward substitution
def bsub(u,y):
    n=len(u)
    x=np.zeros((n, y.shape[1]))
    for i in range(n-1, -1, -1):
        x[i,:]=y[i,:]
        for j in range(i+1, n):
            x[i,:]-=u[i,j]*x[j,:]
        x[i,:]/=u[i,i]
    return x
    
# Determinant calculation
def det(u):
    return np.prod(np.diag(u))
    
# Main execution
l, u = Dec(A)
print("Lower triangular matrix: ", l)
print("Upper triangular matrix: ", u)

# Determinant
print("Determinant of A is: ", det(u))

# Solving AX=V
y=fsub(l, v)
x=bsub(u, y)
ans=np.around(x, decimals=2)
print("Solution (X): ", ans)
```
