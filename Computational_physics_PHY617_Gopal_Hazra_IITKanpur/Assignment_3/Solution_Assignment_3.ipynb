```python
#Question1
N=int(input("Give the value of N : "))
A=np.fromfunction(lambda i, j: i * j/100, (N, N), dtype=float)
B=np.fromfunction(lambda i, j: (i+1) * (j+1)/100, (N, N), dtype=float)
#Using numpy
print("Using Numpy")
print(round(np.sum(np.dot(A, B)),3))
#Using loop
print("Using loop")
C=np.zeros((N,N))
for i in range(len(A[0])):
    for j in range(len(B)):
        C[j][i]=np.sum(A[j,:]*B[:,i])
print(round(np.sum(C),3))
```

```python
#Question 2
x=[0,2,4,6,8,10,12]
t=[1,2,3,4,5,6,7]
v_arr=[]
a_arr=[]
def kinematics(i,x,t):
    v=(x[i+1]-x[i-1])/(t[i+1]-t[i-1])
    a=2*(t[i+1]-t[i-1])**-1*((x[i+1]-x[i])/(t[i+1]-t[i])-(x[i]-x[i-1])/(t[i]-t[i-1]))
    return v,a
def test_kinematics():
    for i in range(1,len(x)-1):
        v,a=kinematics(i,x,t)
        v_arr.append(v)
        a_arr.append(a)
test_kinematics()
print(v_arr,a_arr)

data=np.loadtxt("GPS_data.txt",skiprows=1)
x=data[:,1]
t=data[:,0]
v,a=kinematics(5,x,t)
print(v,a)
```

```python
#Question 3
def count_pairs(dna: str, pair: str) -> int:
    return sum(1 for i in range(len(dna) - 1) if dna[i:i+2] == pair)

# (a)
dna = "ACTGCTATCCATT"
pair = "AT"
print(count_pairs(dna, pair))  # Output: 2

def count_substring(dna: str, substring: str) -> int:
    n = len(substring)
    return sum(1 for i in range(len(dna) - n + 1) if dna[i:i+n] == substring)

#(b)
dna = "ACGTTACGGAACG"
substring = "ACG"
print(count_substring(dna, substring))  # Output: 3
```
